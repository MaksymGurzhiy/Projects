```python
import pandas as pd
import numpy as np

# Загрузка данных
url = 'https://archive.ics.uci.edu/ml/machine-learning-databases/adult/adult.data'
df = pd.read_csv(url, na_values=' ?', header = None)
```

```python
# просматриваем набор данных
print(" Shape:", df.shape)
print("\n Info:")
df.info()

print("\n First rows:")
display(df.head())
```

```python
## устанавливаем название для столбцов
column_names = ['age', 'workclass', 'fnlwgt', 'education', 'education_num',
                'marital_status', 'occupation', 'relationship', 'race', 'sex',
                'capital_gain', 'capital_loss', 'hours_per_week', 'native_country', 'income']
df.columns = column_names

# просматриваем набор данных
print(" Shape:", df.shape)
print("\n Info:")
df.info()

print("\n First rows:")
print(df.head())

```

```python
# доступ к столбцам
print(df['age'])
print(df[['age', 'education']])
print(df.iloc[0:5, 0:3])
print(df['sex'] == 'male')
```

```python
male = df[df['sex'] == ' Male'][['age', 'education']]
display(male)
```

```python
print(df.describe())
```

```python
# 7 Розрахуйте середні значення кількісних ознак в класах
numeric_columns = df.select_dtypes(include=['float64', 'int64']).columns
class_mean = df.groupby('income')[numeric_columns].mean()
print(class_mean)
```

```python
## 8-9 статистикa
for column in ['workclass', 'education', 'marital_status', 'occupation', 'relationship', 'race', 'sex', 'native_country']:
    print(df[column].value_counts())
```

```python
## 10-11
import pandas as pd

# Загрузим данные
df.columns

# Найдем пустые значения
print("Пустые значения до замены:")
print(df.isna().sum())

# Замена пропущенных значений на средние для числовых столбцов
numeric_columns = df.select_dtypes(include=['float64', 'int64']).columns
df[numeric_columns] = df[numeric_columns].apply(lambda x: x.fillna(x.mean()))

# Замена пропущенных значений на моду для категориальных столбцов
categorical_columns = df.select_dtypes(include=['object']).columns
df[categorical_columns] = df[categorical_columns].apply(lambda x: x.fillna(x.mode()[0]))

# Проверка, что пропущенные значения заменены
print("Пустые значения после замены:")
print(df.isna().sum())

```

```python
## 12
from sklearn.preprocessing import LabelEncoder

df = pd.get_dummies(df, columns=['workclass', 'education', 'marital_status', 'occupation', 'relationship', 'race', 'sex', 'native_country'], drop_first=True)
label_encoder = LabelEncoder()
df['income'] = label_encoder.fit_transform(df['income'])
print(df[['age', 'fnlwgt', 'education_num', 'capital_gain', 'capital_loss', 'hours_per_week']])
```

```python
## 13
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
df[['age', 'fnlwgt', 'education_num', 'capital_gain', 'capital_loss', 'hours_per_week']] = scaler.fit_transform(df[['age', 'fnlwgt', 'education_num', 'capital_gain', 'capital_loss', 'hours_per_week']])
print(df[['age', 'fnlwgt', 'education_num', 'capital_gain', 'capital_loss', 'hours_per_week']])
```
